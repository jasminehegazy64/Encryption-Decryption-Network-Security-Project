```python
# Playfair Cipher

# Function to convert the string to lowercase
def toLowerCase(text):
	return text.lower()

# Function to remove all spaces in a string
def removeSpaces(text):
	newText = ""
	for i in text:
		if i == " ":
			continue
		else:
			newText = newText + i
	return newText

# Function to group 2 elements of a string as a list element
def Diagraph(text):
	Diagraph = []
	group = 0
	for i in range(2, len(text), 2): #ba loop etnenat we kol etnen, group them in dig
		Diagraph.append(text[group:i])

		group = i
	Diagraph.append(text[group:])
	return Diagraph
# Function to fill a letter in a string element if 2 letters in the same string matches
def FillerLetter(text):
	k = len(text)
	if k % 2 == 0: #law even fa mafesh letter zyada bs momken yeba feh mm
		for i in range(0, k, 2):
			if text[i] == text[i+1]: # loop on letter in twos
				new_word = text[0:i+1] + str('x') + text[i+1:] #law 2 letters same, make i+1 =x
				new_word = FillerLetter(new_word)
				break
			else:
				new_word = text
	else:
		for i in range(0, k-1, 2):
			if text[i] == text[i+1]:
				new_word = text[0:i+1] + str('x') + text[i+1:]
				new_word = FillerLetter(new_word)
				break
			else:
				new_word = text
	return new_word


list1 = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'k', 'l', 'm',
		'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

# Function to generate the 5x5 key square matrix
def generateKeyTable(word, list1):
	key_letters = []
	for i in word:
		if i not in key_letters:
			key_letters.append(i)

	compElements = []
	for i in key_letters:
		if i not in compElements:
			compElements.append(i)
	for i in list1:
		if i not in compElements:
			compElements.append(i)

	matrix = []   #bazwd letters el fel list, law letter mawgod fel key_letters, skip
	while compElements != []:
		matrix.append(compElements[:5])
		compElements = compElements[5:]

	return matrix


def search(mat, element):
    for i in range(5):
        for j in range(5):
            if mat[i][j] == element:
                return i, j
    # Return a default value if the element is not found
    return -1, -1



def encrypt_RowRule(matr, e1r, e1c, e2r, e2c):
	char1 = ''
	if e1c == 4:
		char1 = matr[e1r][0] #cause its 5x5, fa law ana on 4, means im in last cell, wrap around
	else:
		char1 = matr[e1r][e1c+1] #emshy khatwa

	char2 = ''
	if e2c == 4:
		char2 = matr[e2r][0]
	else:
		char2 = matr[e2r][e2c+1]

	return char1, char2


def encrypt_ColumnRule(matr, e1r, e1c, e2r, e2c):
	char1 = ''
	if e1r == 4:
		char1 = matr[0][e1c]
	else:
		char1 = matr[e1r+1][e1c]  #nafs kalam el rows

	char2 = ''
	if e2r == 4:
		char2 = matr[0][e2c]
	else:
		char2 = matr[e2r+1][e2c]

	return char1, char2


def encrypt_RectangleRule(matr, e1r, e1c, e2r, e2c):
	char1 = ''
	char1 = matr[e1r][e2c]
#emsk akher rohak
	char2 = ''
	char2 = matr[e2r][e1c]

	return char1, char2


def encryptByPlayfairCipher(Matrix, plainList):
    CipherText = []
    for i in range(0, len(plainList)):
        c1 = 0
        c2 = 0
        ele1_x, ele1_y = search(Matrix, plainList[i][0])
        
        # Check if the element is found
        if ele1_x != -1 and ele1_y != -1:
            ele2_x, ele2_y = search(Matrix, plainList[i][1])

            if ele1_x == ele2_x:
                c1, c2 = encrypt_RowRule(Matrix, ele1_x, ele1_y, ele2_x, ele2_y)
            elif ele1_y == ele2_y:
                c1, c2 = encrypt_ColumnRule(Matrix, ele1_x, ele1_y, ele2_x, ele2_y)
            else:
                c1, c2 = encrypt_RectangleRule(Matrix, ele1_x, ele1_y, ele2_x, ele2_y)

            cipher = c1 + c2
            CipherText.append(cipher)
        else:
            # Handle the case where an element is not found
            print(f"Element not found: {plainList[i]}")

    return CipherText


with open(r"C:\Users\USER\Desktop\NetworkSecurity\ALLinput.txt", 'r') as file:
        text_Plain = file.readline()
text_Plain = removeSpaces(toLowerCase(text_Plain))
PlainTextList = Diagraph(FillerLetter(text_Plain))
if len(PlainTextList[-1]) != 2:
	PlainTextList[-1] = PlainTextList[-1]+'z'
key = "Monarchy"
print("Key text:", key)
key = toLowerCase(key)
Matrix = generateKeyTable(key, list1)

print("Plain Text:", text_Plain)
CipherList = encryptByPlayfairCipher(Matrix, PlainTextList)

CipherText = ""
for i in CipherList:
	CipherText += i
#print("CipherText:", CipherText)

output_file_path = r"C:\Users\USER\Desktop\NetworkSecurity\FPoutput.txt"
with open(output_file_path, 'w') as output_file:
        output_file.write(CipherText)
#print(f"CipherText: {CipherText} - Output saved to {output_file_path}\n")
# text_Plain = 'instruments'
# text_Plain = removeSpaces(toLowerCase(text_Plain))
# PlainTextList = Diagraph(FillerLetter(text_Plain))
# if len(PlainTextList[-1]) != 2:
# 	PlainTextList[-1] = PlainTextList[-1]+'z'




```

```python
# Hill Cipher
keyMatrix = [[0] * 3 for i in range(3)]

# Generate vector for the message
messageVector = [[0] for i in range(3)]

# Generate vector for the cipher
cipherMatrix = [[0] for i in range(3)]

# Following function generates thekey matrix for the key string
def getKeyMatrix(key):
    k = 0
    for i in range(3):
        for j in range(3):
            keyMatrix[i][j] = ord(key[k]) % 65
            k += 1

# Following function encrypts the message
def encrypt(messageVector):
    for i in range(3):
        for j in range(1):
            cipherMatrix[i][j] = 0
            for x in range(3):
                cipherMatrix[i][j] += (keyMatrix[i][x] *
                                    messageVector[x][j])
            cipherMatrix[i][j] = cipherMatrix[i][j] % 26

def HillCipher(message, key):

    # Get key matrix from the key string
    getKeyMatrix(key)

    # Generate vector for the message
    for i in range(3):
        messageVector[i][0] = ord(message[i]) % 65

    # Following function generates the encrypted vector
    encrypt(messageVector)

    # Generate the encrypted text from the encrypted vector
    CipherText = []
    for i in range(3):
        CipherText.append(chr(cipherMatrix[i][0] % 26 + 65)) 
    return "".join(CipherText)
        

   


with open(r"C:\Users\USER\Desktop\NetworkSecurity\ALLinput.txt", 'r') as file:
    message_hill = file.readline()

key_hill = "GYBNQKURP"
ciphertext_hill = HillCipher(message_hill, key_hill)

# Check if ciphertext_hill is None
if ciphertext_hill is not None:
    output_file_path = r"C:\Users\USER\Desktop\NetworkSecurity\Hilloutput.txt"
    with open(output_file_path, 'w') as output_file:
        output_file.write(ciphertext_hill)
#     print(f"Message: {message_hill}, Key: {key_hill}, Ciphertext: {ciphertext_hill} - Output saved to {output_file_path}\n")
# else:
#     print("Error: HillCipher returned None.")


```

```python
# Vignere Cipher
def vigenere_cipher(plaintext, key, auto_mode=True):
    # Convert both plaintext and key to uppercase
    plaintext = plaintext.upper()
    key = key.upper()

    # Initialize the ciphertext result
    ciphertext = ""

    # Counter to keep track of the position in the key
    key_position = 0

    # Iterate through each character in the plaintext
    for char in plaintext:
        if char.isalpha():
            # Get the corresponding key character based on the key mode
            if auto_mode:
                key_char = key[key_position % len(key)]
            else:
                key_char = key[key_position % len(key)]

            # Encrypt the character using the Vigenere cipher
            encrypted_char = chr(((ord(char) + ord(key_char) - 2 * ord('A')) % 26) + ord('A'))

            # Update the key position
            key_position += 1

            # Append the encrypted character to the ciphertext
            ciphertext += encrypted_char
        else:
            # Keep non-alphabetic characters unchanged
            ciphertext += char

    return ciphertext

# Test the function with the given keys and modes
with open(r"C:\Users\USER\Desktop\NetworkSecurity\ALLinput.txt", 'r') as file:
        plaintext_vigenere = file.readline()
repeating_key_vigenere = "PIE"
auto_key_vigenere = "AETHER"
ciphertext_repeating = vigenere_cipher(plaintext_vigenere, repeating_key_vigenere, auto_mode=False)
ciphertext_auto = vigenere_cipher(plaintext_vigenere, auto_key_vigenere, auto_mode=True)

output_file_path_repeating = r"C:\Users\USER\Desktop\NetworkSecurity\V2output.txt"
with open(output_file_path_repeating, 'w') as output_file:
    output_file.write(ciphertext_repeating)
#print(f"Repeating Key: {repeating_key_vigenere}, Ciphertext (Repeating Mode): {ciphertext_repeating} - Output saved to {output_file_path_repeating}")

output_file_path_auto = r"C:\Users\USER\Desktop\NetworkSecurity\V2autooutput.txt"
with open(output_file_path_auto, 'w') as output_file:
    output_file.write(ciphertext_auto)
#print(f"Auto Key: {auto_key_vigenere}, Ciphertext (Auto Mode): {ciphertext_auto} - Output saved to {output_file_path_auto}\n")
```

```python
# Vernam Cipher
def vernam_cipher(plaintext, key):
    # Ensure the key is as long as the plaintext
    key = key[:len(plaintext)]

    # Convert characters to ASCII and perform XOR operation
    ciphertext = ''.join(chr(ord(p) ^ ord(k)) for p, k in zip(plaintext, key))

    return ciphertext

with open(r"C:\Users\USER\Desktop\NetworkSecurity\ALLinput.txt", 'r') as file:
        plaintext_vernam = file.readline()
key_vernam = "SPARTANS"
ciphertext_vernam = vernam_cipher(plaintext_vernam, key_vernam)
output_file_path_vernam = r"C:\Users\USER\Desktop\NetworkSecurity\Voutput.txt"
with open(output_file_path_vernam, 'w') as output_file:
    output_file.write(ciphertext_vernam)
# print(f"Plaintext: {plaintext_vernam}")
# print(f"Key: {key_vernam}")
# print(f"Ciphertext: {ciphertext_vernam} - Output saved to {output_file_path_vernam}")
```
